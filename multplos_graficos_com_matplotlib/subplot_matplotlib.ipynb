# Múltiplos Gráficos com  Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("seaborn")

%matplotlib inline
```

# Gráficos em uma mesma figura

Com a função subplots (), você pode desenhar vários gráficos em uma figura:

Nesse primeiro exemplo, vamos plotar dois sub gráficos de linhas 


```python
plt.figure(figsize = ((12, 6)))
plt.subplot(1, 2, 1)
plt.title("Posição do Gráfico  01", fontsize = 16)
plt.text(0.5, 0.5, 'subplot(2,1,1)', ha='center', va='center',
        size=24, alpha=.5)

plt.subplot(1, 2, 2)
plt.text(0.5, 0.5, 'subplot(2,1,2)', ha='center', va='center',
        size=24, alpha=.5)
plt.title("Posição do Gráfico  02", fontsize = 16)

plt.suptitle("Gráficos", fontsize = 20)

plt.show()
```

### Gráficos lado a lado

```python
# Tamanho da imagem principal em polegadas
plt.figure(figsize = ((12, 6)))

#Gráfico 1:
x = np.arange(0, 10, 0.1)
y = np.cosh(x)

plt.subplot(1, 2, 1)
plt.plot(x,y)

#Gráfico 2:
x = np.arange(0, 10, 0.2)
y = np.sin(x)

plt.subplot(1, 2, 2)
plt.plot(x,y)


plt.show()
```

```python
plt.figure(figsize = ((12, 6)))

#Gráfico 1:
x = np.arange(0, 10, 0.1)
y = np.cosh(x)

plt.subplot(1, 2, 1)
plt.boxplot(y)

#Gráfico 2:
x = np.arange(0, 10, 0.2)
y = np.sin(x)

plt.subplot(1, 2, 2)
plt.boxplot(y)


plt.show()
```

### Gráfico um em cima do outro

```python
plt.figure(figsize = ((9, 8)))

#Gráfico 1:
x = np.arange(0, 10, 0.1)
y = np.cosh(x)

plt.subplot(2, 1, 1)
plt.plot(x,y)

#Gráfico 2:
x = np.array([0, 1, 2, 3])
y = np.array([7, 6, 5, 4])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

# Múltiplos Gráficos com Subplot

Você pode plotar quantos gráficos quiser em uma figura, para isso você só precisar dizer a quantidade de linhas e colunas que terá a figura. além do índice do gráfico.


```python
plt.figure(figsize = ((12, 6)))

x = np.arange(0, 10, 0.2)
y = np.tan(x)

plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("Gráfico 01", fontsize = 16)


x = np.arange(0, 10, 0.2)
y = np.sin(x)

plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("Gráfico 02", fontsize = 16)

x = np.arange(0, 10, 0.2)
y = np.sinh(x)

plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title("Gráfico 03", fontsize = 16)

x = np.arange(0, 10, 0.2)
y = np.exp(x)

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title("Gráfico 04", fontsize = 16)

plt.suptitle("Gráficos", fontsize = 20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = ((12, 6)))

x = np.arange(0, 10, 0.2)
y = np.tan(x)

plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title("Gráfico 01", fontsize = 16)


x = np.arange(0, 10, 0.2)
y = np.sin(x)

plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title("Gráfico 02", fontsize = 16)

x = np.arange(0, 10, 0.2)
y = np.sinh(x)

plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title("Gráfico 03", fontsize = 16)

x = np.arange(1, 10, 0.2)
y = np.log(x)

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title("Gráfico 04", fontsize = 16)

plt.suptitle("Gráficos", fontsize = 20)

plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.2, 
                    hspace=0.35)

plt.show()
```
