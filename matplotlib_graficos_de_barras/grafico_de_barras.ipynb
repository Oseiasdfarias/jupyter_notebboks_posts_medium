# Primeiros passos com Matplotlib | Gráfico de Barra

### Hoje vamos aprender como plotar gráficos em barra, para uma melhor visualização dos dados
No Google Colaboratory as bibliotecas Matplotlib e numpy que serão usadas nesse notebbok já são instaladas por padrão.

```python
# Importando a biblioteca
import matplotlib.pyplot as plt

%matplotlib inline
```

## Plotagem na Vertical

```python
# Plotando um gráfico de barras simples
figura = plt.figure()
ax = figura.add_axes([0,0,1.5,1])
series_netflix = ["CoComelon", "Bridgerton", "Ginny & Georgia", "iCarly",
                  "Amigas Para Sempre"]
pontos = [465, 438, 354, 351, 238]
ax.bar(series_netflix, pontos, color = "purple")
ax.set_title("Netflix - Séries Mais Assistidas de 2021", fontsize=16)
ax.set_xlabel("Pontos", fontsize=15)
ax.set_ylabel("Séries", fontsize=15)
plt.show()
```

## Plotagem na Horizontal

```python
# Plotando um gráfico de barras simples
figura = plt.figure()
ax = figura.add_axes([0,0,1.5,1.3])
series_netflix = ["CoComelon", "Bridgerton", "Ginny & Georgia", "iCarly",
                  "Amigas Para Sempre"]
pontos = [465, 438, 354, 351, 238]
ax.barh(series_netflix, pontos, color = "purple")
ax.set_title("Netflix - Séries Mais Assistidas de 2021", fontsize=16)
ax.set_xlabel("Pontos", fontsize=15)
ax.set_ylabel("Séries", fontsize=15)
plt.show()
```
